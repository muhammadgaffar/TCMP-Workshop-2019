```julia
mutable struct systemlattice
    atomicDistance::Float64
    latticeVector::Array{Float64,2}
    numberofatoms::Int64
    description::String
end
```

```julia
graphene = systemlattice(0,zeros(Float64,3,3),0,"sdf")
```

```julia
graphene.atomicDistance = 1.42
graphene.latticeVector = [1 2 3; 4 5 6; 9 8 7]
graphene.numberofatoms = 2
graphene.description = "sistem honeycomb"
```
